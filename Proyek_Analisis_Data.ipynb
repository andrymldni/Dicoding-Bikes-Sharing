### Pertanyaan 2: Apakah pengguna sepeda lebih cenderung keluar saat cuaca cerah atau saat cuaca buruk?

```python
# Menyalin data dari hour_df
weather_plot = hours_df[['weather_condition', 'count']].copy()

# Mengganti data weather_condition dengan kondisi dominannya
weather_plot.replace({
    'weather_condition': {
        1: 'Cerah',
        2: 'Kabut',
        3: 'Hujan Ringan, Salju Ringan',
        4: 'Hujan Lebat, Salju, Kabut'
    }
}, inplace=True)

# Agregat total pengguna berdasarkan weather_condition
weather_plot = weather_plot.groupby('weather_condition')['count'].sum().reset_index()

# Urutkan tabel berdasarkan total pengguna dengan urutan menurun
weather_plot = weather_plot.sort_values('count', ascending=False)

# Definisikan palet warna
color_list = ['#1565c0', '#bbdefb', '#bbdefb', '#bbdefb']

# Buat barplot
fig, ax = plt.subplots(figsize=(9, 5))
sns.barplot(data=weather_plot, x='count', y='weather_condition', palette=color_list)

plt.title("Total Pengguna Sepeda dalam Berbagai Kondisi Cuaca")
plt.ylabel("Cuaca")
plt.xlabel("Total Pengguna")

# Tambahkan data label
for p in ax.patches:
    width = p.get_width()
    plt.text(width, p.get_y() + p.get_height() / 2, f"{width:,.0f}", ha='left', va='center')

plt.tight_layout()
plt.show()
```

### RFM

```python
from datetime import datetime

# Hitung Nilai Recency, Frequency, dan Monetary
hari_ini = datetime.now()
all_df['recency'] = (hari_ini - all_df['date']).dt.days
all_df['frequency'] = all_df['count_x'] + all_df['count_y']
all_df['monetary'] = all_df['count_x'] * all_df['registered_x'] + all_df['count_y'] * all_df['registered_y']

# Kuartil RFM
recency_kuartil = pd.qcut(all_df['recency'], q=4, labels=False)
frequency_kuartil = pd.qcut(all_df['frequency'], q=4, labels=False)
monetary_kuartil = pd.qcut(all_df['monetary'], q=4, labels=False)

# Assign Skor RFM ke DataFrame
all_df['R'] = recency_kuartil + 1  # Ditambah 1 agar skor dimulai dari 1
all_df['F'] = frequency_kuartil + 1
all_df['M'] = monetary_kuartil + 1

# Gabungkan Nilai RFM menjadi satu kolom
all_df['RFM_Score'] = all_df['R'].astype(str) + all_df['F'].astype(str) + all_df['M'].astype(str)

# Tampilkan DataFrame dengan Skor RFM
print(all_df[['date', 'recency', 'frequency', 'monetary', 'R', 'F', 'M', 'RFM_Score']])
```

Kode Python ini menghitung skor RFM untuk pelanggan dan menambahkannya ke DataFrame. Skor RFM dapat digunakan untuk segmentasi pelanggan, kampanye pemasaran, dan analisis churn.

```python
import matplotlib.pyplot as plt

# Visualisasi Skor RFM
plt.figure(figsize=(10, 6))

# Plot untuk RFM Score
all_df['RFM_Score'].value_counts().sort_index().plot(kind='bar', color='lightcoral')
plt.title('RFM Scores Distribution')
plt.xlabel('RFM Scores')
plt.ylabel('Count')

plt.show()
```
